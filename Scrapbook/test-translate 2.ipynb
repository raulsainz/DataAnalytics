```python
#!/usr/bin/python3

"""
    search.py

    MediaWiki API Demos
    Demo of `Search` module: Search for a text or title

    MIT License
"""

import requests
from pprint import pprint
from urllib import parse

S = requests.Session()

URL = "https://en.wikipedia.org/w/api.php"

SEARCHPAGE = "zika virus disease"
#SEARCHPAGE = parse.quote(SEARCHPAGE)
PARAMS = {
    "action": "query",
    "list": "search",
    "format": "json",
    "srsearch": SEARCHPAGE,
    "srqiprofile":'classic',
    "srprop":"snippet"

}


R1 = S.get(url=URL, params=PARAMS)
DATA1 = R1.json()
if('suggestion' in DATA1['query']['searchinfo']):
    print("no encontro resultado cambiando por "+ DATA1['query']['searchinfo']['suggestion'])
    title = DATA1['query']['searchinfo']['suggestion']
else:
    title = DATA1['query']['search'][0]['title']

PARAMS2 = {
    "action": "query",
    "format": "json",
    "prop": "extracts",
    "titles": title,
    "exchars":'500',
    "explaintext":1
}
R2 = S.get(url=URL, params=PARAMS2)
DATA = R2.json()
first_key = list(DATA['query']['pages'].keys())[0]
if 'extract' in DATA['query']['pages'][first_key]:
    print(DATA['query']['pages'][first_key]['extract'])
else:
    print(False)
```
